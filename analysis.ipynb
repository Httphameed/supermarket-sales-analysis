# Supermarket Sales Analysis

This notebook contains an analysis of supermarket sales data from 3 branches over a 3-month period.

## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Loading and exploring the data

```python
df = pd.read_csv('data/supermarket_sales.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

## Data cleaning

```python
df.isnull().sum()
```

No missing values found. Let's convert the date column to datetime format and extract month and day of week for further analysis.

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
df['DayOfWeek'] = df['Date'].dt.day_name()
```

## Data visualization

### Sales by product line

```python
product_line_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=product_line_sales.index, y=product_line_sales.values)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Sales by branch

```python
branch_sales = df.groupby('Branch')['Total'].sum()

plt.figure(figsize=(8, 5))
sns.barplot(x=branch_sales.index, y=branch_sales.values, palette='viridis')
plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

### Customer type distribution

```python
customer_type = df['Customer type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_type.index, y=customer_type.values, palette='pastel')
plt.title('Customer Type Distribution')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### Daily sales trend

```python
daily_sales = df.groupby('Date')['Total'].sum()

plt.figure(figsize=(14, 6))
daily_sales.plot()
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
