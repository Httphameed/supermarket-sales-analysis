# Supermarket Sales Analysis

This is a simple data analysis project using supermarket sales data.

## Step 1: Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Step 2: Load and Explore the Data

```python
# Load the dataset
df = pd.read_csv('data/supermarket_sales.csv')

# Display first 5 rows
print("First 5 rows of the dataset:")
df.head()
```

```python
# Check dataset info
print("\nDataset Info:")
df.info()
```

```python
# Check summary statistics
print("\nSummary Statistics:")
df.describe()
```

## Step 3: Data Cleaning

```python
# Check for missing values
print("Missing values per column:")
df.isnull().sum()
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and day of week for analysis
df['Month'] = df['Date'].dt.month_name()
df['DayOfWeek'] = df['Date'].dt.day_name()
```

## Step 4: Exploratory Data Analysis (EDA)

### 4.1 Sales by Product Line

```python
# Calculate total sales by product line
product_line_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=product_line_sales.index, y=product_line_sales.values)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 4.2 Sales by Branch

```python
# Calculate total sales by branch
branch_sales = df.groupby('Branch')['Total'].sum()

plt.figure(figsize=(8, 5))
sns.barplot(x=branch_sales.index, y=branch_sales.values, palette='viridis')
plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

### 4.3 Customer Type Distribution

```python
# Customer type count
customer_type = df['Customer type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_type.index, y=customer_type.values, palette='pastel')
plt.title('Customer Type Distribution')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### 4.4 Sales Trend Over Time

```python
# Daily sales trend
daily_sales = df.groupby('Date')['Total'].sum()

plt.figure(figsize=(14, 6))
daily_sales.plot()
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Key Findings

1. Food and beverages generate the highest sales
2. Branch C has the highest total sales among all branches
3. Member customers and normal customers are almost equally distributed
4. Sales show some fluctuation over the 3-month period
