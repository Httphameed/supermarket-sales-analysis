# Supermarket Sales Analysis

Hey, this is just some stuff I did with a supermarket sales dataset I found online. Nothing too fancy, just messing around.

First, let's import the libraries I need

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

Okay, let's load the data and see what it looks like

```python
df = pd.read_csv('data/supermarket_sales.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Let's check if there are any missing values

```python
df.isnull().sum()
```

Cool, no missing data. Let's fix the date column and add month/day columns

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month_name()
df['DayOfWeek'] = df['Date'].dt.day_name()
```

Alright, let's make some plots to see what's going on

First, let's see which product lines sell the most

```python
product_line_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=product_line_sales.index, y=product_line_sales.values)
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Okay, food and beverages are top. Now let's check the different branches

```python
branch_sales = df.groupby('Branch')['Total'].sum()

plt.figure(figsize=(8, 5))
sns.barplot(x=branch_sales.index, y=branch_sales.values, palette='viridis')
plt.title('Total Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.show()
```

Branch C is doing the best. Let's see what kind of customers are coming in

```python
customer_type = df['Customer type'].value_counts()

plt.figure(figsize=(8, 5))
sns.barplot(x=customer_type.index, y=customer_type.values, palette='pastel')
plt.title('Customer Type Distribution')
plt.xlabel('Customer Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Pretty even split between members and normal customers. Let's check how sales change over time

```python
daily_sales = df.groupby('Date')['Total'].sum()

plt.figure(figsize=(14, 6))
daily_sales.plot()
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Alright, that's it for now. Just some basic plots to see the trends.
